```python
import transformers
import pandas as pd
```

```python
# load the test data
test = pd.read_csv('test.csv')
test.head()
```

```python
!unzip /content/fine_tuned_model.zip
```

```python


# Predict. 
# Transformers library provide task based pipeline to make our task easy.

from transformers import pipeline
from transformers import AutoTokenizer
# load from previously saved model
pipe = pipeline("text-classification",
                model="/content/content/test_trainer/checkpoint-12500",
                tokenizer=AutoTokenizer.from_pretrained("distilbert-base-cased"))

tokenizer_kwargs = {'padding':True,'truncation':True,'max_length':512}

prediction = pipe('This hotel is awesome',**tokenizer_kwargs)
prediction
```

```python
predictions = []
for r in test['reviews']:
  try:
    pred = pipe(r, **tokenizer_kwargs)[0]['label']
    if pred == 'LABEL_0':
      predictions.append(0)
    else:
      predictions.append(1)


  except:
    predictions.append(0)
    print(r)
```

```python
from sklearn.metrics import classification_report

print(classification_report(test['label'], predictions))
```
